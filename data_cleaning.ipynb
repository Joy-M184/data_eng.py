```python
import pandas as pd

df = pd.read_csv('jobs.csv')

df.head()



```

```python
#standardized date 
df['date']= pd.to_datetime(df['date'], errors = 'coerce')

df.head(10)
```

```python
df.describe()
```

```python
df.info()
```

```python
#salary extraction
#  Clean salary text
df['salary_clean'] = (
    df['salary']
    .str.replace(',', '', regex=False)
    .str.replace('£', '', regex=False)
)

# Extract minimum salary (first number)
df['min_salary'] = df['salary_clean'].str.extract(r'(\d+)').astype(float)

#  Extract maximum salary (second number if exists, else same as min)
df['max_salary'] = df['salary_clean'].str.extract(r'\d+\s*to\s*(\d+)').astype(float)
df['max_salary'] = df['max_salary'].fillna(df['min_salary'])

#  Get overall min + max
overall_min = df['min_salary'].min()
overall_max = df['max_salary'].max()

```

```python
df.drop(columns=['salary'], inplace=True)
df.drop(columns=['salary_clean'], inplace=True)
```

```python
df.head()
```

```python
import pandas as pd
import re

# Extract postcode using UK postcode pattern
postcode_pattern = r'([A-Z]{1,2}\d[A-Z\d]?\s*\d[A-Z]{2})'
df['postcode'] = df['location'].str.extract(postcode_pattern)

# Extract city name (everything before the first comma)
df['location'] = df['location'].str.split(',').str[0].str.strip()

```

```python
df.head()
```

```python
# columns rearrangement
df=df[['title', 'company','date','location','postcode','min_salary','max_salary']]

df.head()
```

```python
df.isnull().sum()
```

```python
# saved the cleaned data
df.to_csv('cleaned_jobs.csv')
```
